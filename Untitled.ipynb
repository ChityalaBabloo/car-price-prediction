<p><b>How do I covnvert my categorical values into some other values. for that we will be using getdummies, a function inside the pandas. A technique which uses one hot encoding. this is done as follows:</b></p>

```python
final_dataset = pd.get_dummies(df,drop_first=True)
#drop_first means removing the first value in each level
#suppose if a car is of petrol, them it means, it is not disel, not cng. which means 1 0 0
#similarly, 010 ---> diesel; 001--->cng
final_dataset.head()
```

### Now, let's split the training data and testing data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.25)
```

```python
X_train.shape
```

I'm using Randomforest Regression -- so, when I use this, I don't need to scale the values because it uses decision tree, usually in decision tree it is not required

```python
from sklearn.ensemble import RandomForestRegressor
rf_random = RandomForestRegressor()
```

```python
#Hyperparamters -- basically using differnt different values for the parameters
import numpy as np
n_estimators = [int(x) for x in np.linspace(start=100,stop=1200,num=12)]
n_estimators #decision trees
```

```python
#Randomized search CV helps to find the best parameters out of the these
n_estimators = [int(x) for x in np.linspace(start=100,stop=1200,num=12)]
#Number of features to consider at every split
max_features = ['auto','sqrt']
#Max nuo.of levels in tree
max_depth = [int(x) for x in np.linspace(5,30,num=6)]
#Minimum num of samples rquired to split a node
min_samples_split = [25,10,15,100]
#minimum number of samples required at each leaf node
min_samples_leaf=[1,2,5,10]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestRegressor()
```

```python
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
#As our model is created, let's predict
predictions = rf_random.predict(X_test)
predictions
```

```python
sns.distplot(y_test-predictions)
```
